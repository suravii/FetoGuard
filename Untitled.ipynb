```python
class_1 = data[data['fetal_health'] == 1]
class_2 = data[data['fetal_health'] == 2]
class_3 = data[data['fetal_health'] == 3]

# Muestrear aleatoriamente la mitad de los datos de la clase 1
class_1_sampled = resample(class_1,
                           replace=False,  # sin reemplazo
                           n_samples=len(class_1) // 2,  # la mitad de los datos
                           random_state=42)  # para reproducibilidad

# Combinar los datos muestreados con todas las clases minoritarias
df_balanced = pd.concat([class_1_sampled, class_2, class_3])

# Verificar la distribuciÃ³n de las clases en el conjunto de datos equilibrado
print(df_balanced['fetal_health'].value_counts())
```

```python
X=df_balanced.drop(["fetal_health"],axis=1)
y=df_balanced["fetal_health"]

col_names = list(X.columns)
X_df = pd.DataFrame(X, columns=col_names)

# Split data into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.3, random_state=42, stratify=y)

# Scaling datasets
s_scaler = StandardScaler()
X_train= s_scaler.fit_transform(X_train)
X_test = s_scaler.transform(X_test)

X_train= pd.DataFrame(X_train, columns=col_names)
X_test= pd.DataFrame(X_test, columns=col_names)

# Verify distribution of classes in each dataset
print("Training dataset distribution of classes:")
print(y_train.value_counts())
print("\nTesting dataset distribution of classes:")
print(y_test.value_counts())
```

```python
forest = RandomForestClassifier(n_estimators=200, max_depth=None, random_state=42)
forest.fit(X_train, y_train)

importances = forest.feature_importances_
feature_names = np.array([c for c in data.columns if c != 'fetal_health'])
forest_importances = pd.Series(importances, index=feature_names)
std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)

fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")

most_important_features = feature_names[np.argsort(forest_importances)[-10:]]
print("Most important features: ", most_important_features)
```

# Selecting the model

```python
# Naive Bayes
# Build model
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# Calculate metrics
metrics = classification_report(y_test, y_pred)
kappa = cohen_kappa_score(y_test, y_pred)

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Binarize labels for ROC curves
y_test_bin = label_binarize(y_test, classes=[1, 2, 3])
y_probs = clf.predict_proba(X_test)

# Calculate ROC curves and AUC for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(3):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Create figure with subplots
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Confusion matrix subplot
sns.set(font_scale=1.2)
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', ax=ax[0])
ax[0].set_xlabel('Predicted label')
ax[0].set_ylabel('True label')
ax[0].set_title('Confusion Matrix')

# ROC curves subplot
for i in range(3):
    ax[1].plot(fpr[i], tpr[i], label='ROC curve class {0} (area = {1:0.2f})'.format(i+1, roc_auc[i]))
ax[1].plot([0, 1], [0, 1], 'k--')
ax[1].set_xlim([0.0, 1.0])
ax[1].set_ylim([0.0, 1.05])
ax[1].set_xlabel('False Positive Rate')
ax[1].set_ylabel('True Positive Rate')
ax[1].set_title('Receiver Operating Characteristic')
ax[1].legend(loc="lower right")

# Show figure
plt.tight_layout()
plt.show()

print("")
print("Classification Report:")
print(metrics)
print("Kappa Score:", kappa)
```

```python
# KNN
# Define hyperparameter ranges for grid search
param_grid = {
    'n_neighbors': list(range(1, 31)),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

# Create KNN model
knn = KNeighborsClassifier()

# Create a custom metric that prioritizes class 3 and assigns less weight to class 2
# We use average = 'weighted' to account for class imbalance
# beta = 2 indicates that recall is given twice the priority compared to precision
def custom_scorer(y_true, y_pred):
    score_class_3 = fbeta_score(y_true, y_pred, beta=2, average='weighted', labels=[3])
    score_class_2 = fbeta_score(y_true, y_pred, beta=2, average='weighted', labels=[2])
    return 0.66 * score_class_3 + 0.33 * score_class_2

scorer = make_scorer(custom_scorer, greater_is_better=True)

# Run grid search
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring=scorer)
grid_search.fit(X_train, y_train)

# Show best hyperparameters
print("Best hyperparameters:", grid_search.best_params_)

# Create KNN model with best hyperparameters
best_knn = grid_search.best_estimator_
best_knn.fit(X_train, y_train)

# Predict with optimized model
y_pred = best_knn.predict(X_test)

# Calculate and show confusion matrix
cm = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Subplot 1: Confusion matrix
plt.subplot(1, 2, 1)
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues')
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix')

# Calculate and show ROC curves for each class
y_test_bin = label_binarize(y_test, classes=[1, 2, 3])
y_probs = best_knn.predict_proba(X_test)
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(3):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Subplot 2: ROC curves
plt.subplot(1, 2, 2)
for i in range(3):
    plt.plot(fpr[i], tpr[i], label='ROC curve class {0} (area = {1:0.2f})'.format(i+1, roc_auc[i]))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic for Optimized KNN')
plt.legend(loc="lower right")

plt.tight_layout()
plt.show()

# Calculate and show metrics
kappa = cohen_kappa_score(y_test, y_pred)
print("")
print(classification_report(y_test, y_pred))
print("Custom score:", custom_scorer(y_test, y_pred))
print("Kappa Score:", kappa)
```
